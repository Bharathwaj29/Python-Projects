```python
#List of items in the store
userName=input("enter your name: ")
welcomeMessage=f"Welcome to my store {userName}"
lenWMsg=len(welcomeMessage)
print("*"*lenWMsg)
print(welcomeMessage)
print()

store_items={
               'Milkshake:200ml': {'quantity':10,'price':50.23},
               'Apple:Pack of 4': {'quantity':20,'price':99.0}, 
               'Chicken-750gms': {'quantity':10,'price':200.0}, 
               'Orange:Pack of 5': {'quantity':15,'price':85.99}, 
               'Mango:Pack of 4': {'quantity':10,'price':120.99}, 
               'Coffee:200gms': {'quantity':15,'price':249.99}, 
               'Milk:500ml':{'quantity':30,'price':29.99}, 
               'Curd:200ml':{'quantity':20,'price':20.0}
                }
                                   
#function to display items
def display_items():
    print("\nItems available in store: ")
    print("Item\t\t\tQuantity\tPrice")
    for item, details in store_items.items():
        print(f"{item}\t{details['quantity']}\t\t{details['price']}")
                                   
#entering the number of quantities required
def get_customer_order():
    customer_order={}
    for item, details in store_items.items():                               
        quantity=int(input(f"\nHow many {item} do you want?"))
        if quantity > details['quantity']:
            print(f"Sorry we have {details['quantity']} {item}(s) in stock")                        
        else:
            customer_order[item]={'quantity':quantity, 'price':details['price']}                       
            details['quantity']-=quantity
    return customer_order

                                   
#calculate delivery charge
def calculate_delivery_charge(distance):
     if distance<=5:
        return 10
     elif distance<=10:
        return 12
     elif distance<=15:
        return 14
     else:                              
        return 20
                                   
 #Final billing of choosen items and customer details
def display_bill(customer_order,name,address,distance):
    total = 0
    print("\nItems purchased")
    print("\nItem\t\t\t\tQuantity\tPrice")                             
    for item, details in customer_order.items():
        print(f"{item}\t\t{details['quantity']}\t\t{details['price']}")                         
        total+=details['quantity']*details['price']                         
    delivery_charges=calculate_delivery_charge(distance)                             
    total+=delivery_charges                             
    print(f"\nDelivery charges:{delivery_charges}")
    print(f"\nTotal bill amount:{total}")
    print(f"\nDeelivering to:{name} at {address}")
                                   
                                   
 #loop to keep program running
while True:
    #display list of items available
    display_items()      
    #get customer order                             
    customer_order=get_customer_order()
    #Get customer details      
    name=input("\nEnter your name: " )      
    address=input("Enter your address: ")      
    distance=float(input("Enter the distance: "))      
    #generating final bill      
    display_bill(customer_order, name, address, distance)      
    #asking user to continue shopping or not
    choice=input("Do you want to continue shopping: (Yes/No)")
    if choice=='No':
        print("**********Thankyou for shopping************")
        print("Hope see you")    
        break

            
            
            







```
