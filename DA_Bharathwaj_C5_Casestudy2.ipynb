```python
def separate_data_types(credit_df):
    categorical=[]
    continuous=[]
    for column in credit_df.columns:
        if credit_df[column].nunique() <25:
            categorical.append(column)
        else:
            continuous.append(column)
    return categorical, continuous

categorical, continuous=separate_data_types(credit_df)

from tabulate import tabulate
print(tabulate({"categorical":categorical,
                 "continuous":continuous},
                 headers=["categorical", "continuous"]))
```

### Handling Outliers :

```python
plt.figure(figsize=(8,4))
sns.boxplot(credit_df.AMT_INCOME_TOTAL);
```

```python
q1=credit_df['AMT_INCOME_TOTAL'].describe()['25%']
q3=credit_df['AMT_INCOME_TOTAL'].describe()['75%']
iqr=q3-q1

upper_bound=q3+1.5*iqr
credit_df['AMT_INCOME_TOTAL']=np.where(credit_df['AMT_INCOME_TOTAL']>upper_bound, upper_bound, 
                                       credit_df['AMT_INCOME_TOTAL'])
```

```python
plt.figure(figsize=(8,4))
sns.boxplot(credit_df.AMT_INCOME_TOTAL);
```

### We can see that outliers are removed from above graph.
